**Importando os dados e blibiotecas**


```
import pandas as pd
import seaborn as sns
```

```
candidatos_PR = pd.read_csv("/content/drive/My Drive/Data_Eleições/consulta_cand_2020_PR.csv", sep=';', encoding='iso-8859-1'); 
```

```
candidatos_PR
```

Filtrar candidatos de londrina.

```
candidatos_LONDRINA = candidatos_PR[candidatos_PR.NM_UE == 'LONDRINA']
candidatos_LONDRINA
```

Verificar candiatos que irão se reeleger.

```
candidatos_LONDRINA['ST_REELEICAO'].value_counts(normalize=True)*100
```

```
candidatos_LONDRINA[candidatos_LONDRINA['ST_REELEICAO'] == 'S']
```

**Analisar o GRAU de escolaridade dos candidatos de Londrina**

```
candidatos_LONDRINA['DS_GRAU_INSTRUCAO'].value_counts(normalize=True)*100
```

```
candidatos_LONDRINA['SG_PARTIDO'].value_counts(normalize=True)*100
```

Análise de escolaridade por partido.

```
candidatos_LONDRINA[candidatos_LONDRINA['SG_PARTIDO']=='PT']['DS_GRAU_INSTRUCAO'].value_counts(normalize=True).plot(kind='bar', title='Grau de Instrucao Candidatos 2020');
```

```
candidatos_LONDRINA[candidatos_LONDRINA['SG_PARTIDO']=='PC do B']['DS_GRAU_INSTRUCAO'].value_counts(normalize=True).plot(kind='bar', title='Grau de Instrucao Candidatos PC do B 2020');
```

```
candidatos_LONDRINA[candidatos_LONDRINA['SG_PARTIDO']=='PSL']['DS_GRAU_INSTRUCAO'].value_counts(normalize=True).plot(kind='bar', title='Grau de Instrucao Candidatos PSL 2020');
```
